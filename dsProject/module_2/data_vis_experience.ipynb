#### 1. Reproducing a plot
1. Use `numpy`'s `random` module to generate a random list of size 1000 of the following three categories: "CS", "Engineering", "Math" and "Physics"
  1.1 Use a seed of 42
2. Store the data in an array called my `students_affiliations`
3. use `matplotlib` to plot the data in `students_affiliations` to reproduce the following graph
 3.1 The plot uses "#FF5733" as color with an alpha (transparency) of 0.55.

* Your graph should look exactly like the one below

![](https://www.dropbox.com/s/ffmu62es6d7evh5/experience_fig.png?dl=1)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

1. Get 1000 random choices from the four areas below, and turn into a series.
2. For each of the four areas, calculate how many of them there are.
3. Put all the four values together into an array
4. Turn the data into a dataframe, having counts in the first column, and disciplines in the second.
5. Sort the values by the counts with the bigger ones on the top.
6. Plot the graph using the values of counts and disciplines extracted from the sorted dataframe.
7. Add the labels (x and y-axis),and the title.

```python
random.seed(42)
students_affiliations = pd.Series(random.choices(["Engineering", "Math", "Physics", "CS"], k=1000))
```

```python
cs_count = students_affiliations.value_counts()["CS"]
eng_count = students_affiliations.value_counts()["Engineering"]
math_count = students_affiliations.value_counts()["Math"]
phy_count = students_affiliations.value_counts()["Physics"]
affiliations_count = [eng_count, math_count, phy_count, cs_count]
```

```python
df = pd.DataFrame({'Counts':affiliations_count,'Disciplines':['Engineering','Math', 'Physics','CS']})
sorted_df = df.sort_values(by='Counts', ascending = False)
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot()
_ = plt.bar(sorted_df['Disciplines'], sorted_df['Counts'], color='#FF5733', alpha=0.55)
ax.set_title("Distribution of Students in the Program", color='green')
ax.set_xlabel("Disciplines")
ax.set_ylabel("Counts")
```

#### 2. Generating a scatter plot
* Generate random data to produce a scatter plot
  * You can use any function to generate the data
  * Instead of using circles as symbols, you should use squares.
    

1. Get 50 random age values from 1-18 (cats), 1-16 (dogs).
2. Get 50 random expense values from 220-700 (cats), 200-800 (dogs).
3. Get 50 random choices from the array [0, 120, 360] which will be the unexpected medical expense yearly.
4. Make the scatter plot using those values, and add labels.

```python
plt.figure(figsize=(12,6))

age_cats = np.random.randint(1, 18, 50)
age_dogs = np.random.randint(1, 16, 50)
expense_cats = np.random.randint(220, 700, 50)
expense_dogs = np.random.randint(200, 800, 50)
medical_expense = np.random.choice([0, 120, 360], 50)

plt.scatter(age_cats, expense_cats, label= "Cat Expense", s=medical_expense, marker='s')
plt.scatter(age_dogs, expense_dogs, label= "Dog Expense", s=medical_expense, marker='s')

plt.legend()

plt.xlabel("Age")
plt.ylabel("Expense")
_ = plt.title("Representation of cat and dogs' age and expense")
```
